### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()


print("Ventas por categoría Tienda 1:")
display(ventas_por_categoria_tienda1)
```

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por categoría Tienda 1:")
display(ventas_por_categoria_tienda1)

print("\nVentas por categoría Tienda 2:")
display(ventas_por_categoria_tienda2)

print("\nVentas por categoría Tienda 3:")
display(ventas_por_categoria_tienda3)

print("\nVentas por categoría Tienda 4:")
display(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos_tienda1 = tienda['Producto'].value_counts().head()
productos_menos_vendidos_tienda1 = tienda['Producto'].value_counts().tail()

productos_mas_vendidos_tienda2 = tienda2['Producto'].value_counts().head()
productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts().tail()

productos_mas_vendidos_tienda3 = tienda3['Producto'].value_counts().head()
productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts().tail()

productos_mas_vendidos_tienda4 = tienda4['Producto'].value_counts().head()
productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts().tail()


print("Productos más vendidos Tienda 1:")
display(productos_mas_vendidos_tienda1)
print("\nProductos menos vendidos Tienda 1:")
display(productos_menos_vendidos_tienda1)

print("\nProductos más vendidos Tienda 2:")
display(productos_mas_vendidos_tienda2)
print("\nProductos menos vendidos Tienda 2:")
display(productos_menos_vendidos_tienda2)

print("\nProductos más vendidos Tienda 3:")
display(productos_mas_vendidos_tienda3)
print("\nProductos menos vendidos Tienda 3:")
display(productos_menos_vendidos_tienda3)

print("\nProductos más vendidos Tienda 4:")
display(productos_mas_vendidos_tienda4)
print("\nProductos menos vendidos Tienda 4:")
display(productos_menos_vendidos_tienda4)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4}")
```
